```python
import pandas as pd
```

```python
df = pd.read_csv("/content/x_learner_training.csv")
```

```python
DISCOUNT_COST = 100
AVG_REVENUE = 1000
```

```python
df.columns
```

```python
df["expected_gain"] = df["uplift_pred"] * AVG_REVENUE
df["roi"] = df["expected_gain"] - DISCOUNT_COST
```

```python
total_profit = df["roi"].sum()
print(total_profit)
```

```python
df.groupby("segment")["roi"].mean()
```

```python
df[df["roi"] > 0].sort_values("roi", ascending=False)
```

```python
def decision(uplift):
    if uplift > 0.2:
        return "Offer Discount"
    elif uplift > 0:
        return "Low Priority Offer"
    else:
        return "Do Not Target"
```

```python
df["Decision"] = df["uplift_pred"].apply(decision)
```

```python
df.to_csv("customer_cate_predictions.csv", index=False)
```
